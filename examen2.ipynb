```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import numpy as np
from PIL import Image
from scipy.ndimage import gaussian_gradient_magnitude
from wordcloud import WordCloud, ImageColorGenerator
```

```python
url = 'https://www.senamhi.gob.pe/?&p=pronostico-meteorologico'
page = requests.get(url)
page.status_code
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
lugar0 = []
for l in soup.select("#listaprono > table > tbody  tr > td > div:nth-child(1)"):
    lugar0.append(l.get_text())
lugar = []
for l in lugar0:
    for i in range(3):
        lugar.append(l)
```

```python
#tabla de 3 filas 

fecha = []
for f in soup.select("#listaprono > table  tbody > tr > td div > div:nth-child(1)"):
    fecha.append(f.get_text())
```

```python
tmax = []
for t in soup.select("#listaprono > table  tbody > tr > td div > div:nth-child(3)"):
    tmax.append(t.get_text())
```

```python
tmin = []
for t in soup.select("#listaprono > table  tbody > tr > td div > div:nth-child(4)"):
    tmin.append(t.get_text())
```

```python
descripcion = []
for d in soup.select("#listaprono > table  tbody > tr > td div > div:nth-child(5)"):
    descripcion.append(d.get_text())
```

```python
df = pd.DataFrame({'Lugar':lugar,'Fecha':fecha,'Tmax':tmax,'Tmin':tmin,'Descripcion':descripcion})
df
```

```python
df['Tmed'] = [ (int(x[:-2])+int(n[:-2]))/2 for x,n in zip(df['Tmax'],df['Tmin'])]
df
```

```python
import hashlib

df['hash256_desc'] = [hashlib.sha256(d.encode('utf-8')).hexdigest() for d in df['Descripcion']]
```

```python
#reordenando
df = df[['Lugar', 'Fecha', 'Tmax', 'Tmin', 'Tmed', 'hash256_desc']]
df
```

```python
df.to_excel('df2.xlsx',sheet_name='Sheet1')
```

```python
import os 

class Banco:
    def __init__(self,nombre):
        self.nombre = nombre
        self.cola = []
        self.clientes = []
    def llegada(self,data):
        if data['preferencia'] == 1:
            self.cola.insert(0,data)
        else:
            self.cola.append(data)
    def getCola(self):
        return self.cola
    def atenderCliente(self):
        if (len(self.cola)) == 0:
            print('no existen mas clientes')
        else:
            print('atendiendo ..')
            os.system('sleep 0.5')
            print('cliente atendido y eliminado de la cola \n',self.cola[0])
            # una persona sera cliente una vez es atendido.
            self.clientes.append(self.cola[0])
            ## eliminando al cliente de la cola 
            del self.cola[0]
    def buscarCliente(self,dni):
        # buscando solamente a clientes que ya se atendieron almenos una vez.
        #utilizando busqueda secuencial
        for cliente in self.clientes:
            if cliente['DNI'] == dni:
                return cliente
                break
        return "cliente no encontrado"
```

```python
banco_cpp = Banco('banco de la nacion')
## ingresando valores manualmente

banco_cpp.llegada({'Apellidos': 'yataco ramirez', 'Nombres': 'luis alberto', 'DNI': 33, 'preferencia': 1})
banco_cpp.llegada({'Apellidos': 'luyo takayama', 'Nombres': 'eduardo', 'DNI': 44, 'preferencia': 0})
banco_cpp.llegada({'Apellidos': 'calderon maldonado', 'Nombres': 'patricia rossana', 'DNI': 331, 'preferencia': 1})

##
while True:
    print("""
    • 1 = Llegada de cliente
    • 2 = Llegada de cliente preferencial
    • 3 = Atención de cliente
    • 4 = Búsqueda de cliente
    • 5 = Limpiar consola
    • 0 = Fin de eventos
    """)
    opcion = int(input("> "))
    if opcion in [1,2]: 
        #pidiendo los datos al cliente.
        data = {
            'Apellidos':None,
            'Nombres':None,
            'DNI':None,
        }
        for key in data.keys():
            print("ingrese %s :" %key,end='')
            data[key] = input()
        data['DNI'] = int(data['DNI'])
        data['preferencia'] = opcion - 1 
        banco_cpp.llegada(data)
        print('\nCola actualizada:')
        for c in banco_cpp.getCola():
            print(c)
        print()
    elif opcion == 3:
        banco_cpp.atenderCliente()

    elif opcion == 4:
        dni = int(input('ingrese un dni: '))
        resultado = banco_cpp.buscarCliente(dni)
        print(resultado)

    elif opcion == 5:
        os.system('clear')

    elif opcion == 5:
        break
    elif opcion == 0:
        break
    else:
        pass
```
